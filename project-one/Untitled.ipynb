```python
import pandas as pd

# Sample data
data = {'DATE': ['2017-01-01','2018-01-01','2019-01-01', '2020-01-01', '2021-01-01', '2022-01-01'],
        'FLPOP': [20977.089,21254.926,21492.056, 21589.602, 21828.069, 22244.823]}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert the 'DATE' column to datetime
df['DATE'] = pd.to_datetime(df['DATE'])

# Extract the year component and create a new 'Year' column
df['Year'] = df['DATE'].dt.year

# Drop the original 'DATE' column if it's no longer needed
df = df.drop(columns=['DATE'])

# Print the updated DataFrame
print(df)
```

```python
import matplotlib.pyplot as plt
# Assuming you have a DataFrame named df with 'Year' and 'FLPOP' columns
years = df['Year']
population = df['FLPOP']
plt.figure(figsize=(10, 6))
plt.plot(years, population, marker='o', linestyle='-')
plt.title('Population Trend Over the Years')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.xticks(years)
plt.show()
```

```python
import pandas as pd

# Sample data
data = {'DATE': ['2017-01-01','2018-01-01','2019-01-01', '2020-01-01', '2021-01-01', '2022-01-01'],
        'Median Income': [62600,63160,66480, 65110, 64410, 65370]}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert the 'DATE' column to datetime
df['DATE'] = pd.to_datetime(df['DATE'])

# Extract the year component and create a new 'Year' column
df['Year'] = df['DATE'].dt.year

# Drop the original 'DATE' column if it's no longer needed
df = df.drop(columns=['DATE'])

# Print the updated DataFrame
print(df)
```

```python
years = df['Year']
population = df['Median Income']
plt.figure(figsize=(10, 6))
plt.plot(years, population, marker='o', linestyle='-')
plt.title('Income Trends')
plt.xlabel('Year')
plt.ylabel('Median Income')
plt.grid(True)
plt.xticks(years)
plt.show()
```
